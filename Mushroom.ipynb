```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mushroom=pd.read_csv('mushroom.csv',na_values='?')
```

```python
mushroom
```

```python
mushroom.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy='most_frequent')
```

```python
mushroom['stalk-root']=si.fit_transform(mushroom['stalk-root'].values.reshape(-1,1))
```

```python
mushroom.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#PREPROCESSING
```

```python
encoder=LabelEncoder()
for j in mushroom.columns:
    mushroom[j]=encoder.fit_transform(mushroom[j].astype(str))
```

```python
mushroom
```

```python
mushroom.columns
```

```python
#splitting the data in x and y.
x=mushroom.drop('class',axis=1)
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x=scaler.fit_transform(x)
```

```python
x
```

```python
x=pd.DataFrame(data=x)
x
```

```python
y=mushroom['class']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score 
```

```python
list=[LogisticRegression(),GaussianNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier()]
for j in list:
    r_max=0
    print('for the algorithm: ',j)
    for i in range(33,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
        j.fit(x_train,y_train)
        pred=j.predict(x_test)
        accuracy=accuracy_score(pred,y_test)
        print('accuracy score corresponding to random state',i,'is: ',accuracy)
        if accuracy>r_max:
            r_max=accuracy
            final=i
    print('\n')
    print('Max accuracy score corresponding to: ', final,'is: ',r_max)
    print('\n')
```

1. Logisitic Regression gave us an accuracy score of 0.9745 at random state 54.
2. GaussianNB gave us an accuracy score of 0.9175 at random state 61.
3. SVC gave us an accuracy score of 1 at random state 33.
4. Decision tree classifier gave us an accuracy score of 1 at random state 33.
5. K neighbors classifier gave us an accuracy score of 1 at random state 33.
6. Random forest classifier gave us an accuracy score of 1 at random state 33.

```python
#Checking for underfitting and overfitting
from sklearn.model_selection import cross_val_score
list=[LogisticRegression(),GaussianNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier()]
for i in list:
    cvscore=cross_val_score(i,x,y,cv=9)
    print('The cross validation score for',i,'is: ',cvscore)
    print('The mean and standard deviations are: ',cvscore.mean(),cvscore.std())
    print('\n')
```

From cross validation score it is visible that our model is working well.

```python
# We will now proceed with Decision tree classifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
print(dtc.score(x_train,y_train))
pred=dtc.predict(x_test)
print('Accuracy Socre is: ',accuracy_score(pred,y_test))
print('Confusion matrix is : ',confusion_matrix(pred,y_test))
print('Classificatin report is: ',classification_report(pred,y_test))
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
y_pred_prob=dtc.predict_proba(x_test)[:,-1]
y_pred_prob
```

```python
fpr,tpr,thresholds=roc_curve(y_test,y_pred_prob)
```

```python
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='Decision Tree Classifier')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('DecisionTree Classifier')
plt.show()
```

```python
auc_score=roc_auc_score(y_test,dtc.predict(x_test))
print(auc_score)
```
