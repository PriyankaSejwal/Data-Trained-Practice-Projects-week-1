Presence of outliers can be observed from the plots above.

```python
#Dropping cetain columns that have high negative correlation with class.
sonar_new=sonar.drop(['V10','V11','V12','V45','V48','V49'],axis=1)
```

```python
sonar_new
```

```python
sonar_new.skew()
```

```python
#Removing skewsness
import numpy as np
from sklearn.preprocessing import PowerTransformer
```

```python
pt=PowerTransformer(method='yeo-johnson',standardize=True)
pt.fit_transform(sonar_new)
```

```python
sonar_new.skew()
```

```python
from scipy.stats import zscore
```

```python
#Removing outliers
z=np.abs(zscore(sonar_new))
```

```python
sonar_data=sonar_new[(z<3).all(axis=1)]
```

```python
sonar_data
```

```python
sonar_data.skew()
```

```python
#Splitting the data into input and output
x=sonar_data.iloc[:,0:-1]
```

```python
x
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=20)
```

```python
x=pca.fit_transform(x)
```

```python
x.shape
```

```python
pd.DataFrame(data=x)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x=sc.fit_transform(x)
```

```python
x
```

```python
x=pd.DataFrame(x)
```

```python
x
```

```python
x.shape
```

```python
y=sonar_data.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score 
```

```python
list=[LogisticRegression(),GaussianNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier()]
for j in list:
    r_max=0
    print('for the algorithm: ',j)
    for i in range(33,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
        j.fit(x_train,y_train)
        pred=j.predict(x_test)
        accuracy=accuracy_score(pred,y_test)
        print('accuracy score corresponding to random state',i,'is: ',accuracy)
        if accuracy>r_max:
            r_max=accuracy
            final=i
    print('\n')
    print('Max accuracy score corresponding to: ', final,'is: ',r_max)
    print('\n')
```

From above calculations, it is observed that SVC is performing better than the other models at random state 91 with accuracy score of 0.9607.

```python
from sklearn.model_selection import cross_val_score
list=[LogisticRegression(),GaussianNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier()]
for i in list:
    cvscore=cross_val_score(i,x,y,cv=9)
    print('The cross validation score for',i,'is: ',cvscore)
    print('The mean and standard deviations are: ',cvscore.mean(),cvscore.std())
    print('\n')
```

##### It is seen that in Cross validation score method, Support vector classifier performed well with average accuracy 0.69948.

```python
#Working on the best model- SVC
svc=SVC(kernel='rbf')
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=91)
svc.fit(x_train,y_train)
print(svc.score(x_train,y_train))
pred=svc.predict(x_test)
print("Accuracy score: ",accuracy_score(y_test,pred))
print("Confusion matrix: ",confusion_matrix(y_test,pred))
print('Classification report: ',classification_report(y_test,pred))
```

```python
import joblib
joblib.dump(svc,'sonar_svc.obj')
svc_load=joblib.load('sonar_svc.obj')
svc_load.predict(x_test)
```
